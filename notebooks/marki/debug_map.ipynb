```python
import dash
from dash import dcc
from dash import html
from dash.dependencies import Input, Output, State
import plotly.graph_objects as go
import plotly.express as px

import warnings

import json

import pandas as pd
import geopandas as gpd
from pandas.errors import SettingWithCopyWarning
import numpy as np

import os
from pathlib import Path

import sys

## Declare app
app = dash.Dash(__name__)

## Prepare data

## Load predictions
df_imp = pd.read_parquet("../../data/pm2_df.parquet")

## Load actual
df_actual = pd.read_parquet("../../data/apart_df.parquet")

gdf_municipalities = gpd.read_file("../../data/municipalities_geo.gpkg").to_crs(
    "EPSG:4326"
)
gdf_municipalities = gdf_municipalities[["tgid", "namedut", "geometry"]]
```

```python
gdf_municipalities["geometry"] = (
    gdf_municipalities.to_crs(gdf_municipalities.estimate_utm_crs())
    .simplify(100)
    .to_crs(gdf_municipalities.crs)
)
```

```python
gdf_municipalities = gdf_municipalities.set_index("namedut")
```

```python
gdf_municipalities["lambda"] = np.random.normal(size=gdf_municipalities.shape[0])
```

```python
fig = px.choropleth(
    gdf_municipalities,
    geojson=gdf_municipalities.geometry,
    locations=gdf_municipalities.index,  # assuming 'namedut' is a column in gdf_municipalities that matches the 'id' in your GeoJSON
    color="lambda",  # 'column' should be the column in your dataframe you want to color your map by
)

fig.update_geos(fitbounds="locations")

fig.show()
```
